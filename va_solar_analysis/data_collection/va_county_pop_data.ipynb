This file gathers a list of all the counties and cities within Virginia, along with their population and area. The info has been cross referenced with other population statistics, as well as the Virginia Department of Elections.

```python
import pandas as pd
from IPython.core.display import HTML

# get the url
url = 'https://en.wikipedia.org/wiki/List_of_cities_and_counties_in_Virginia'

# turn the url into a list
list_of_tables = pd.read_html(url)

# loop through the list enumerated, printing the index and table info
for i, table in enumerate(list_of_tables):
    print(f'table number: {i}')
    print(f'table itself: {table}')

# assign the correct table to a variable
my_table = list_of_tables[2]

# turn that variable to a csv
my_table.to_csv('../csv_collection/county_population_hist_info.csv', index=False)
```

Now I'll verify the csv was created successfully:

```python
va_history_population_df = pd.read_csv('../csv_collection/county_population_hist_info.csv')

print(va_history_population_df.info())
display(HTML(va_history_population_df.to_html()))
```
