This file gathers some of the most important data of the project. It compiles data from a wikipedia entry on the outcome of the 2020 Presidential Election in Virginia by locality. The data has been crossreferenced with data from the Virginia Department of Elections.

```python
import pandas as pd
from IPython.core.display import HTML

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/2020_United_States_presidential_election_in_Virginia"

# Extract tables from the page
tables = pd.read_html(url)

# Find the correct table (check by printing tables)
for i, table in enumerate(tables):
    print(f"Table {i}:")
    print(table.head())

# The correct table is at index 32
df = tables[32]

# Save to CSV
df.to_csv("../csv_collection/virginia_2020_election.csv", index=False)
```

```python
# Convert the csv to a DataFrame, so it can be inspected for accuracy:
va_election_2020_df = pd.read_csv('../csv_collection/virginia_2020_election.csv')

print(va_election_2020_df.info())
display(HTML(va_election_2020_df.to_html()))
```
